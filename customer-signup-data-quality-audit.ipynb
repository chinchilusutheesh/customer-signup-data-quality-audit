Standardization

```python
df['gender'].unique()
```

```python
df['gender'] = df['gender'].str.strip().str.title()
gender_map = {
    'Female':'Female', 'F':'Female',
    'Male':'Male', 'M':'Male',
    'Non-Binary':'Non-Binary', 'Non Binary':'Non-Binary',
    'Other':'Other'
}

df['gender'] = df['gender'].replace(gender_map)
df['gender'] = df['gender'].fillna('Unknown')
```

```python
df['gender'] = df['gender'].replace({'123':'Unknown'})
```

```python
df['region'] = df['region'].str.strip().str.title()
valid_regions = ['North','South','East','West','Central']
df['region'] = df['region'].where(df['region'].isin(valid_regions), 'Unknown')
```

```python
df['region'].unique()
```

```python
df['plan_selected'] = df['plan_selected'].str.strip().str.title()
plan_map = {
    'Basic':'Basic',
    'Premium':'Premium',
    'Pro':'Pro',
    'Unknownplan':'Unknown'
}

df['plan_selected'] = df['plan_selected'].replace(plan_map)
valid_plans = ['Basic','Premium','Pro','Unknown']
df['plan_selected'] = df['plan_selected'].where(df['plan_selected'].isin(valid_plans), 'Unknown')
```

```python
tickets = pd.read_csv("support_tickets.csv")
tickets.head()

```

```python
tickets.info()
tickets.describe()
```

```python
tickets['ticket_date']= pd.to_datetime(tickets['ticket_date'])
```

```python
df_merged = pd.merge(df, tickets, on='customer_id', how='left')
```

```python
df_merged
```

```python
df_merged.columns
```

```python
df_merged['signup_date'] = pd.to_datetime(df_merged['signup_date'], errors='coerce')
df_merged['ticket_date'] = pd.to_datetime(df_merged['ticket_date'], errors='coerce')
df_merged['days_to_support'] = (df_merged['ticket_date'] - df_merged['signup_date']).dt.days

# Filter customers who raised tickets in 14 days (0â€“14 days)
early_support = df_merged[(df_merged['days_to_support'] >= 0) &
                          (df_merged['days_to_support'] <= 14)]

early_support_count = early_support['customer_id'].nunique()
early_support_count
```

```python
df_merged.groupby(['region', 'plan_selected'])['ticket_id'].count().unstack(fill_value=0)
```

```python
support_summary = df_merged.groupby(['plan_selected', 'region'])['ticket_id'].count().unstack(fill_value=0)
support_summary
```

```python
support_rate = (
    df_merged.groupby(['plan_selected','region'])['ticket_id'].count() /
    df_merged.groupby(['plan_selected','region'])['customer_id'].count()
) * 100

support_rate.unstack()
```
